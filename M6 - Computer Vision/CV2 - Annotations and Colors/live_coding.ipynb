```python
# video

video = cv2.VideoCapture(1)

while(video.isOpened()):
    check, frame = video.read()
    if frame is not None:
        img = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        #img = frame
        cv2.imshow('frame',img)        
        if cv2.waitKey(30) & 0xFF == ord('q'):
            break
    else:
        break


video.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```
